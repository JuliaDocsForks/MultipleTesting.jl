```julia
using MultipleTesting
using StatsBase
using Distributions
using Gadfly
```

```julia
pv = [rand(Uniform(),1000);rand(Beta(0.5,7), 1000)]
println("Storey with λ=0.5: ", estimatepi0(pv, StoreyEstimator(0.5)))
println("Storey bootstrap: ", estimatepi0(pv, StoreyBootstrapEstimator()))
```

Use modified grenander estimator 

```julia
fdrfit = fit(pv, GrenanderEstimator(StoreyEstimator(0.5)))

plot(x=pvalues(fdrfit), y=distribution(fdrfit), Geom.line)
```

```julia
plot(x=pvalues(fdrfit), y=tailfdr(fdrfit), Geom.line)
```

```julia
plot(x=pvalues(fdrfit), y=localfdr(fdrfit), Geom.line)
```
